```python
# This is a sample Python script.

# Press Shift+F10 to execute it or replace it with your code.
# Press Double Shift to search everywhere for classes, files, tool windows, actions, and settings.


 # Press Ctrl+F8 to toggle the breakpoint.
import  pandas as pd
import  numpy as np


# Press the green button in the gutter to run the script.
from pip._vendor import chardet

if __name__ == '__main__':
    with open("HW1data.csv",'rb') as a:
        result=chardet.detect(a.read(10000))

    print(result)
    #编码修改
    data_row = pd.read_csv("HW1data.csv", encoding='ISO-8859-1')
    data_row = pd.DataFrame(data_row)
     #显示修改
    pd.set_option('display.max_columns', 100000)

    pd.set_option('display.width', 100000)

    pd.set_option('display.max_colwidth', 10000)
    #成功读入数据，发现所有数据列10%左右的缺失
    for j in data_row.columns:
            b=data_row[j].isnull().sum()/len(data_row[j])*100
            print(round(b,2))
        #PatientId去除重复的
    data_row.dropna(axis=0, how="any",subset=['PatientId'], inplace=True)
    data_row.reset_index(inplace=True, drop=True)

    deep_data=['PatientId','Gender','Age','Neighbourhood']
    data_2a=data_row[deep_data]  #提取的数据
   
    
    data_row.dropna(axis=0,how="any",subset=['ScheduledDay', 'AppointmentDay','SMS_received','No-show'],inplace=True)
    
    data_row.drop_duplicates(subset=['PatientId'],keep='first',inplace=True)
    
    
    data_row.reset_index(inplace=True,drop=True)

#补数据
    for i in range(len(data_row)):
        for j in ['Gender','Age','Neighbourhood']:
               #if i%100==0:
                #   print(i)#显示进度，当为604091就跑完
               if pd.isnull(data_row.loc[i,j]):
                   c=data_2a[(data_2a['PatientId']==data_row.loc[i,'PatientId'])]
                   c.reset_index(inplace=True, drop=True)
                   for k in range(c.shape[0]):
                       if pd.isnull(c.loc[k,j])==False:
                           data_row.loc[i, j]=c.loc[k,j]
                           break
                   

  
# 去没不全数
    data_row.dropna(axis=0,how="any",subset=['PatientId','Gender','Age','Neighbourhood','ScheduledDay','AppointmentDay','SMS_received','No-show'],inplace=True)
    data_row.fillna({"Scholarship":0,"Hipertension":0,"Diabetes":0,"Alcoholism":0,"Handcap":0},inplace=True)

    data_row.reset_index(inplace=True,drop=True)

#删除列：
    data_row.drop("PatientId",axis=1,inplace=True)
    data_row.drop("AppointmentID",axis=1,inplace=True)

#找出有问题的age
    total=data_row.shape[0]
    for i in range(total):
       if data_row.loc[i,'Age']<0 or data_row.loc[i,'Age']>150:
          data_row.drop(labels=i,axis=0,inplace=True)

    data_row.reset_index(inplace=True,drop=True)
    
    print('please wait minutes, this is very fast')
#计算scheduledDay 和AppointmentDay的差距天数
    data_row['ScheduledDay']=pd.to_datetime(data_row['ScheduledDay'])
    data_row['AppointmentDay']=pd.to_datetime(data_row['AppointmentDay'])

    del_day=(pd.to_datetime(data_row['ScheduledDay']) - pd.to_datetime(data_row['AppointmentDay'])).dt.days

    data_row.insert(0,'Delta_Da',del_day,allow_duplicates=True)


#负数去除
    total=data_row.shape[0]
    for i in range(total):
        if data_row.loc[i,'Delta_Da']<0:
           data_row.drop(labels=i, axis=0, inplace=True)
    
    data_row.reset_index(inplace=True,drop=True)

#计算星期几
    A_week=data_row['AppointmentDay'].dt.dayofweek+1
    S_week=data_row['ScheduledDay'].dt.dayofweek+1
    data_row.insert(0,"SDay_DOW",S_week,allow_duplicates=True)
    data_row.insert(0,"ADay_DOW",A_week,allow_duplicates=True)

    
#去除数据
    data_row.drop("AppointmentDay",axis=1,inplace=True)
    data_row.drop("ScheduledDay",axis=1,inplace=True)

#打印处理的结果
    print(data_row)
    print("end")



```
